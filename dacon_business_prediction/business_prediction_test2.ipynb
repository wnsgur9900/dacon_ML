```python
import pandas as pd
```

```python
train = pd.read_csv("data-files/train.csv")
test = pd.read_csv("data-files/test.csv")
submission = pd.read_csv("data-files/sample_submission.csv")
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
except_cols = ["ID", "성공확률"]
X_train = train.drop(columns=except_cols)
X_test = test.drop(columns = "ID")
y_train = train["성공확률"]
```

```python
cat_cols = ["인수여부", "상장여부", "국가", "투자단계"]
```

```python
X_train = pd.get_dummies(X_train, columns=cat_cols)
X_test = pd.get_dummies(X_test, columns=cat_cols)
```

```python
X_train.isnull().sum()
```

```python
X_train['직원 수'] = X_train['직원 수'].fillna(X_train['직원 수'].mean())
X_train['고객수(백만명)'] = X_train['고객수(백만명)'].fillna(X_train['고객수(백만명)'].mean())
X_train['분야'] = X_train['분야'].fillna(0)
X_train['기업가치(백억원)'] = X_train['기업가치(백억원)'].fillna(0)

X_test['직원 수'] = X_test['직원 수'].fillna(X_test['직원 수'].mean())
X_test['고객수(백만명)'] = X_test['고객수(백만명)'].fillna(X_test['고객수(백만명)'].mean())
X_test['분야'] = X_test['분야'].fillna(0)
X_test['기업가치(백억원)'] = X_test['기업가치(백억원)'].fillna(0)
```

```python
X_train['설립연도'] =train['설립연도'].astype('object')
X_test['설립연도'] =test['설립연도'].astype('object')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
cols = ["기업가치(백억원)", "분야"]
X_train = pd.get_dummies(X_train, columns=cols)
X_test = pd.get_dummies(X_test, columns=cols)
```

```python
X_train
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
submission = pd.DataFrame({
    'ID': test['ID'],             # test 데이터에 있는 ID 컬럼
    '성공확률': y_pred
})

submission['성공확률'] = submission['성공확률'].round(3)
```

```python
submission.to_csv("data-files/submission_RandomForest3.csv",index=False)
```
